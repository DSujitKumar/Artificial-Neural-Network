```python
# Regression based problems
# General equstion  y=3x^2+5

# Basic Libraries to be imported
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl
%matplotlib inline
```

```python
min_val=-20
max_val=20
num_point=140
x=np.linspace(min_val,max_val,num_point)
```

```python
y=3*np.square(x)+5
y/=np.linalg.norm(y)
```

```python
data=x.reshape(num_point,1)
labels=y.reshape(num_point,1)
plt.scatter(data,labels)
plt.xlabel("Dimention 1")
plt.ylabel("Dimention 2")
plt.title("Data Points")
```

```python
mlnn=nl.net.newff([[min_val,max_val]],[10,6,1])
mlnn.trainf=nl.train.train_gd
```

```python
error_progress=mlnn.train(data,labels,epochs=2000,show=100,goal=0.01)
```

```python
output=mlnn.sim(data)
y_pred=output.reshape(num_point)

x_dence=np.linspace(min_val,max_val,num_point*2)
y_dence_pred=mlnn.sim(x_dence.reshape(x_dence.size,1)).reshape(x_dence.size)
plt.plot(x_dence,y_dence_pred,"-",x,y,".",x,y_pred,"p")
plt.title("Actual vs Predict")
```
